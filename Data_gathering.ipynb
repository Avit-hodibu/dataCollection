````python
#since site has many pages so we need to do in loop to extract all data

df_api_all=pd.DataFrame()
page_number=5
for i in range(1,page_number):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df_api_all = df_api_all._append(temp_df,ignore_index=True)
````

````python
#beautifulSoup is lib use for web scraping
soup=BeautifulSoup(webpage,'lxml')
#This help in traverse in the html code
````

````python
# arrange the content in html structure
print(soup.prettify())
````

````python
# this will give all the p tag which has class name rating from the given page
soup.find_all('p', class_='rating')

# like this
soup.find_all('a', class_='review-count').text
````

````python
company= soup.find_all('div', class_='company-content-wrapper')
name =[]
rating=[]
reviews=[]
ctype=[]
hq=[]
how_old=[]
no_of_employee=[]
for i in company:
  #find means single
  #.text give value in text
  #.strip will strip space
  name.append(i.find('h2').text.strip())
  rating.append(i.find('p', class_='rating').text.strip())
  reviews.append(i.find('a', class_='review-count').text.strip())
  # [0] because when p class is call with find_all it will provide 4 entity
  # because they have use same class style
  #so, when [0] it will only call 1st index value
  ctype.append(i.find_all('p', class_='infoEntity')[0].text.strip())
  # similarly [1] means 2nd index value
  hq.append(i.find_all('p', class_='infoEntity')[1].text.strip())
  how_old.append(i.find_all('p', class_='infoEntity')[2].text.strip())
  no_of_employee.append(i.find_all('p', class_='infoEntity')[3].text.strip())
````

````python
# create dataframe
dict_data = {'name':name, 'rating':rating, 'reviews':reviews, 'ctype':ctype, 'hq':hq, 'how_old':how_old, 'no_of_employee':no_of_employee}
df=pd.DataFrame(dict_data)
````
